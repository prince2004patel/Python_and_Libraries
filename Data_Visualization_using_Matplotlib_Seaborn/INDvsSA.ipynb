```python
# India Bowling: Economy Rate
plt.figure(figsize=(10,6))
bars = plt.barh(ind_bowling_df["Bowler"],ind_bowling_df["Econ"],color="skyblue")

for bar in bars:
    width = bar.get_width()
    plt.text(width,bar.get_y() + bar.get_height()/2,'{:.2f}'.format(width),ha= "left",va="center",color="black",fontsize=10)

plt.title("India Bowling: Economy Rate")
plt.xlabel("Economy Rate")
plt.ylabel("Bowlers")
plt.grid(True)
plt.show()
```
